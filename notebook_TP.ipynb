```python
url = "http://www.ebi.ac.uk/Tools/webservices/psicquic/intact/webservices/current/search/query/pubid:17446270"

try:
    httpReq = requests.get(url, proxies=None)
except NameError:
    httpReq = requests.get(url)
PMID = httpReq.text
print(PMID)
```

```python
import pandas as pd

PMID = pd.read_csv(PMID,header=None, sep="\t")

```

```python

list1=[]
list2=[]
listfinal=[]
list1=PMID.split("\n")
for element in list1 :
    a = element.split("\t")
    list2.append(a)
for x in list2 :
    if x[0].startswith("uniprotkb:") and x[1].startswith("uniprotkb:"):
        listfinal.append( [ x[0].replace("uniprotkb:", ""),  x[1].replace("uniprotkb:", "") ]  + x[2:] )
print(listfinal)
```

```python
print(len(listfinal))
```

```python
liste_virus_virus=[]
liste_virus_human=[]
for x in listfinal:
    if "virus" in x[10] and "virus" in x[9]:
        liste_virus_virus.append(x)
    else :
        liste_virus_human.append(x)
    
```

```python
print(len(liste_virus_human))
print(len(liste_virus_virus))
```

```python
%matplotlib nbagg
import networkx as nx
import matplotlib.pyplot as plt

plt.figure()
plt.axis('off')


G = nx.Graph()
G.add_edge("a", "b")
G.add_edge("a", "c")
for n in G.nodes:
    print(n)

pos = nx.spring_layout(G)
print(pos)

nx.draw_networkx_nodes(G, pos, node_color="blue", node_shape="o", nodelist=["a", "b"] )
nx.draw_networkx_nodes(G, pos, node_color="red",  node_shape="o", nodelist=["c"] )
#nx.draw_networkx_labels(G,pos,font_weight=800,font_color='black')
nx.draw_networkx_edges(G, pos, width=0.5)

# Convert names into richer labels and eventually mask some nodes
labelDict = { 
        "a" : "Milou",
        "b" : "Haddock"#, 
#        "c" : "Tryphon"
        }

#Now only add labels to the nodes you require (the hubs in my case)
nx.draw_networkx_labels(G, pos, labelDict, font_size=16, font_color='r')
```

```python
G2=nx.Graph()
labelDict={}
for element in liste_virus_virus:
    G2.add_edge(element[0],element[1])
    labelDict.update({element[0] : element[0],
             element[1] : element[1]})
pos = nx.spring_layout(G2)

nx.draw_networkx_nodes(G2, pos, node_color="blue", node_shape="o", nodelist=G2.nodes, node_size=10)

nx.draw_networkx_edges(G2, pos, width=0.5)
nx.draw_networkx_labels(G2, pos, labelDict, font_size=6, font_color='r')
```
